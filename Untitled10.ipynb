```python
# Merge the dataframes on pincode
merged_df = pd.merge(og_pune_rera, df_pune_selected, left_on='location_pin_code', right_on='Pincode', how='left')

# Display the head of the merged dataframe
display(merged_df.head())
```

```python
merged_df.isnull().sum()
```

```python
merged_df.drop(columns=["Pincode"],inplace=True,errors="ignore")
```

```python
# Filter rows where 'DivisionName' is null
null_division_rows = merged_df[merged_df['DivisionName'].isnull()]

# Display the location_pin_code for these rows
display(null_division_rows[['location_pin_code', 'DivisionName']])
```

```python
# Get unique location_pin_code values from the filtered rows
unique_null_division_pincodes = null_division_rows['location_pin_code'].unique()

# Display the unique pincodes
print("Unique location pin codes with null DivisionName:")
display(unique_null_division_pincodes)
```

```python
# Filter all_pincodes to get information for the unique null division pincodes
pincode_info = all_pincodes[all_pincodes['Pincode'].isin(unique_null_division_pincodes)]

# Display the relevant columns
print("Division, Office, and Region names for the unique pincodes with null DivisionName:")
display(pincode_info[['Pincode', 'DivisionName', 'OfficeName', 'RegionName']])
```

```python
# Merge merged_df with pincode_info to get the correct division, office, and region names
merged_df_filled = pd.merge(merged_df, pincode_info[['Pincode', 'DivisionName', 'OfficeName', 'RegionName']],
                            left_on='location_pin_code', right_on='Pincode', how='left', suffixes=('', '_correct'))

# Fill the null values in the original columns using the correct values from the merge
for col in ['DivisionName', 'OfficeName', 'RegionName']:
    merged_df_filled[col].fillna(merged_df_filled[col + '_correct'], inplace=True)

# Drop the redundant columns from the merge
merged_df_filled.drop(columns=['Pincode', 'DivisionName_correct', 'OfficeName_correct', 'RegionName_correct'], inplace=True, errors='ignore')

# Display the head of the updated dataframe
display(merged_df_filled.head())
```

```python
# Filter for projects with at least one booked apartment or plot
projects_with_sales = merged_df[(merged_df['number_of_booked_appartments'] > 0) | (merged_df['number_of_plots_booked_/_alloted_/_sold'] > 0)].copy()

# Group by DivisionName and sum the booked counts
bookings_by_division_with_sales = projects_with_sales.groupby('DivisionName')[['number_of_booked_appartments', 'number_of_plots_booked_/_alloted_/_sold']].sum()

# Display the result
display(bookings_by_division_with_sales)
```
